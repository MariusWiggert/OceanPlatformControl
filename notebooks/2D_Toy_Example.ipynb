```python
# This is a sample where we construct a simple, linearly time-varying 2D flow and compare a couple of actuation techniques
```

```python
# Step 1: construct the problem velocity field
v_current = 1.

cur_time_steps = 5

# construct grid
xgrid = np.linspace(-5, 5, 20)
ygrid = np.linspace(0, 12, 20)
t_grid = np.linspace(0, 30, cur_time_steps)
# X, Y = np.meshgrid(xgrid, ygrid, indexing='ij')
u_field = np.zeros((cur_time_steps, 20, 20))
for i in range(cur_time_steps):
    # print(v_current * (1-i/float(cur_time_steps)))
    u_field[i, :, 6:14] = v_current * (1-i/float(cur_time_steps))

    
# make interpolation a differentiable casadi function
u_field_flat = u_field.ravel(order='F')
u_curr_func = ca.interpolant('timeVarCurs', 'bspline', [t_grid, xgrid, ygrid], u_field_flat)

# parcels needs at least two velocity fields as inputs as it always assumes time-varying, hence we create a 3D tensor & time vector
field_time_vec = np.linspace(0, 30, cur_time_steps)
u_field_in_time = np.moveaxis(u_field, [0], [2])
v_field_in_time = np.zeros(u_field_in_time.shape)

# create the field set
data_dict = {'U': u_field_in_time, 'V': v_field_in_time}
dimensions = {'lon': xgrid, 'lat': ygrid, 'time': field_time_vec}
fieldset = p.FieldSet.from_data(data_dict, dimensions, transpose=True, mesh='flat', allow_time_extrapolation=True)
```

## 2) Optimal control: use a non-linear optimizer to find the best path

```python
def plot_sol_variables(sol, N):
    # Plot the results
    plt.figure(1)
    plt.plot(np.linspace(0., sol.value(T), N+1), sol.value(x)[0, :], '--')
    plt.plot(np.linspace(0., sol.value(T), N+1), sol.value(x)[1, :], '-')
    plt.plot(np.linspace(0., sol.value(T), N+1), sol.value(x)[2, :], '-')
    plt.plot(np.linspace(0., sol.value(T), N), sol.value(u)[0, :], '-.')
    plt.plot(np.linspace(0., sol.value(T), N), sol.value(u)[1, :], '-.')
    plt.title('Time series of values of decision variables')
    plt.xlabel('time')
    plt.ylabel('value')
    plt.legend(['x trajectory', 'y trajectory', 'b trajectory', 'u trajectory', 'h trajectory'])
    # plt.grid()
    plt.show()

def plot_x_y_traj(sol):
    # Plot the x-y trajectory
    plt.figure(1)
    plt.plot(sol.value(x)[0, :], sol.value(x)[1, :], '--')
    plt.title('X-Y Trajectory')
    plt.xlim(-5, 5.0)
    plt.ylim(0, 10.0)
    plt.xlabel('x')
    plt.ylabel('y')
    # plt.grid()
    plt.show()
```

## a) Time-optimal with time as variable

```python
opti = ca.Opti()
# settings
N = 100 # Number of control intervals
# recharging settings
c_recharge = 0.2

# declare decision variables
T = opti.variable()
# x, y, b, t
x = opti.variable(4, N+1)   # Decision variables for state trajetcory
u = opti.variable(2, N)
x_start = opti.parameter(4, 1)  # Parameter (not optimized over)
x_goal = opti.parameter(2, 1)  # Parameter (not optimized over)

# shortest time until goal
opti.minimize(T)
```

```python
# specify the dynamics of the system as a function
F = ca.Function('f', [x, u],
                [ca.vertcat(u[0]*ca.cos(u[1]) + u_curr_func(ca.vertcat(x[3], x[0], x[1])),
                            u[0]*ca.sin(u[1]),
                            c_recharge - u[0]**2,
                            1)],
                ['x', 'u'], ['x_dot'])

# add the dynamics constraints (RK4 version)
dt = T/N
for k in range(N):
    k1 = F(x=x[:,k], u=u[:,k])
    k2 = F(x=x[:,k]+ dt/2*k1['x_dot'], u=u[:,k])
    k3 = F(x=x[:,k]+ dt/2*k2['x_dot'], u=u[:,k])
    k4 = F(x=x[:, k] + dt*k3['x_dot'], u=u[:,k])
    x_next = x[:, k] + dt/6*(k1['x_dot']+2*k2['x_dot']+2*k3['x_dot']+k4['x_dot'])
    opti.subject_to(x[:,k+1] == x_next)
    
# boundary constraint
opti.subject_to(T >= 0)
opti.subject_to(x[:4, 0] == x_start)
opti.subject_to(ca.dot(x[:2, -1] - x_goal, x[:2, -1] - x_goal) <= 0.1)

# control constraints
opti.subject_to(opti.bounded(0, u[0, :], 1))
opti.subject_to(opti.bounded(0, u[1, :], 2*ca.pi))
# state constraints
opti.subject_to(opti.bounded(-5., x[0, :], 5))
opti.subject_to(opti.bounded(0., x[1, :], 10.))
opti.subject_to(opti.bounded(0.1, x[2, :], 1.))

# inital values for optimizer
x_init = np.vstack([np.zeros(N+1), np.linspace(0, 10, N+1), 0.5*np.ones(N+1), np.linspace(0, 30, N+1)])
u_init = np.array([[0.5]*N, [np.pi/2]*N])
```

```python
# set parameters for optimization
opti.set_value(x_start, [0, 0, 1, 0])
opti.set_value(x_goal, [0, 10])
# initial values
opti.set_initial(x, x_init)
opti.set_initial(u, u_init)

# run optimizer
opti.solver('ipopt')
sol = opti.solve()
```

```python
plot_sol_variables(sol, N)
```

```python
plot_x_y_traj(sol)
```

```python
# extract the time vector and control signal
T = sol.value(T)
u = sol.value(u)
dt = sol.value(dt)

time_vec = np.arange(u.shape[1]+1) * dt
# create the vehicle as a particle
pset = p.ParticleSet.from_list(fieldset=fieldset,   # the fields on which the particles are advected
                             pclass=particles.OpenLoopParticle,  # the type of particles (JITParticle or ScipyParticle)
                             lon=[0.],    # a vector of release longitudes
                             lat=[0.],   # a vector of release latitudes
                            control_traj=[u],
                            control_time=[time_vec],
                            v_max=[1.])
```

```python
output_file = pset.ParticleFile(name="open_loop_control.nc", outputdt=timedelta(seconds=0.5)) # the file name and the time step of the outputs
open_loop_actuation = pset.Kernel(kernels.open_loop_control)
pset.execute(p.AdvectionRK4 + open_loop_actuation,                 # the kernel (which defines how particles move)
             runtime=timedelta(seconds=T),    # the total length of the run
             dt=timedelta(seconds=1.),      # the timestep of the kernel
             output_file=output_file)
```

## Energy-optimal with a fixed time horizon & no-currents at all

```python
opti = ca.Opti()
# settings
T = 20 # Time horizon
N = 40 # Number of control intervals

# recharging settings
c_recharge = 0.2
```

```python
# Declare variables
x_in = ca.MX.sym('x', 3)	# states (x, y, b)
u_in = ca.MX.sym('u', 2)	# Control (u, h)

# Differential equation
f_x = ca.vertcat(u_in[0]*ca.cos(u_in[1]), u_in[0]*ca.sin(u_in[1]), c_recharge - u_in[0]**2)

# Create an integrator
dae = {'x': x_in, 'p': u_in, 'ode':f_x}
opts = {'tf': T/N} # interval length
intg = ca.integrator('intg', 'rk', dae, opts)

# evaluate with symbols and make a function out of it (x,u)->(x_next)
res = intg(x0=x_in, p=u_in)
F = ca.Function('f', [x_in, u_in], [res['xf']], ['x', 'u'], ['x_next'])

# test the function
# F(x=[0,0,0], u=[1,0])

# integrate N times
sim = F.mapaccum(N)
```

```python
x = opti.variable(3, N+1)   # Decision variables for state trajetcory
u = opti.variable(2, N)
x_start = opti.parameter(3, 1)  # Parameter (not optimized over)
x_goal = opti.parameter(2, 1)  # Parameter (not optimized over)

# minimum energy
opti.minimize(u[0, :] @ u[0, :].T)
for k in range(N):
    opti.subject_to(x[:, k+1] == F(x[:, k], u[:, k]))

# boundary constraint
opti.subject_to(x[:, 0] == x_start)
opti.subject_to(ca.dot(x[:2, -1] - x_goal, x[:2, -1] - x_goal) <= 0.1)

# control constraints
opti.subject_to(opti.bounded(0, u[0, :], 1))
opti.subject_to(opti.bounded(0, u[1, :], 2*ca.pi))
# state constraints
opti.subject_to(opti.bounded(-5., x[0, :], 5))
opti.subject_to(opti.bounded(0., x[1, :], 10.))
opti.subject_to(opti.bounded(0.1, x[2, :], 1.))

# init
x_init = np.vstack([np.zeros(N+1), np.linspace(0, 10, N+1), 0.5*np.ones(N+1)])
u_init = np.array([[0.5]*N, [np.pi/2]*N])
```

```python
# set parameters for optimization
opti.set_value(x_start, [0, 0, 1])
opti.set_value(x_goal, [0, 10])
# initial values
opti.set_initial(x, x_init)
opti.set_initial(u, u_init)

# run optimizer
opti.solver('ipopt')
sol = opti.solve()
```
