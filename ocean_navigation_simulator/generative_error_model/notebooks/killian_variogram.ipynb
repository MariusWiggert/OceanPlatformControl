```python
import os
import pandas as pd
import xarray as xr
import numpy as np
import itertools

root = "/home/jonas/Downloads/Archive/test_download_fc_march_august"
files = os.listdir(root)

data = xr.open_dataset(os.path.join(root, files[0]))
print(data)

# when sampling throw away NaN values
lon_len = len(data["longitude"].values)
lat_len = len(data["latitude"].values)
time_len = len(data["time"].values)
total_len = lon_len * lat_len * time_len

idx = np.random.choice(list(range(total_len)), size=30000)

lon = data["longitude"].values.reshape(-1)
lat = data["latitude"].values.reshape(-1)
time = data["time"].values.reshape(-1)

axes = np.array(list(itertools.product(lon, lat, time)))[idx]
lon = axes[:, 0]
lat = axes[:, 1]
time = axes[:, 2]
u_error = data["utotal"].values.reshape(-1)[idx]
v_error = data["vtotal"].values.reshape(-1)[idx]

data = pd.DataFrame({"lon": lon,
                     "lat": lat,
                     "time": time,
                     "u_error": u_error,
                     "v_error": v_error})

data = data.dropna()
print(data)
```
